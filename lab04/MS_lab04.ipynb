## Theory

### Bayes' Theorem

Naive Bayes classifiers are based on **Bayes' Theorem**, which provides a way to calculate conditional probabilities. The theorem is stated as:

$$ P(class|data) = \frac{P(data|class) \times P(class)}{P(data)} $$

Where:
- $P(class|data)$ is the **posterior probability** - the probability of the class given the observed data
- $P(data|class)$ is the **likelihood** - the probability of observing the data given the class
- $P(class)$ is the **prior probability** - the initial probability of the class before observing data
- $P(data)$ is the **marginal likelihood** - the probability of the data (often ignored in classification since it's constant for all classes)

### Naive Assumption

The "naive" in Naive Bayes comes from the **conditional independence assumption**: all features are assumed to be independent of each other given the class. In reality, features often depend on each other, but this assumption drastically simplifies computation.

For example, in spam detection, the words "free", "money", and "win" might appear together, but Naive Bayes treats each word's contribution independently.

Under this assumption, the likelihood becomes:

$$ P(data|class) = \prod_{i=1}^{n} P(feature_i|class) $$

This makes the algorithm computationally efficient and allows it to work well even with small datasets.

In this lab we implement **Gaussian Naive Bayes**, which models each $P(feature_i|class)$ as a normal distribution. Other variants (Multinomial, Bernoulli) exist for discrete data.

### Estimating Parameters from Data

For each feature in each class, we need a mean $\mu$ and standard deviation $\sigma$ to define its Gaussian distribution. These are estimated directly from the training data using the **sample mean** and **sample standard deviation** (with Bessel's correction):

$$ \hat{\mu} = \frac{1}{n}\sum_{i=1}^n x_i \qquad\qquad s = \sqrt{\frac{1}{n-1}\sum_{i=1}^n (x_i - \hat{\mu})^2} $$

This corresponds to `np.mean()` and `np.std(ddof=1)` in NumPy. The $n{-}1$ denominator corrects for the bias of the maximum-likelihood estimator on small samples.

### Comparison with Other Algorithms

| Feature | Naive Bayes | Logistic Regression | SVM | Decision Tree | Random Forest |
|---------|-------------|--------------------|-----|---------------|--------------|
| **Assumptions** | Feature independence | Linear relationship | No assumptions | No assumptions | No assumptions |
| **Speed** | Very Fast | Fast | Medium/Slow | Fast | Medium |
| **Small Data** | Excellent | Poor (overfitting) | Good | Fair | Good |
| **Feature Correlation** | Struggles | Handles well | Handles | Handles | Handles |
| **Model Complexity** | Simple | Simple | Complex | Medium | Complex |
| **Interpretability** | Moderate | High (coefficients) | Low | High (structure) | Low |
| **Scalability** | Excellent | Good | Fair (kernel methods) | Fair | Good |

#### When to use Naive Bayes:
- Text classification (spam, sentiment analysis, topic classification)
- Real-time applications where speed is critical
- Multiclass classification problems
- Small datasets with limited training examples

#### When to avoid Naive Bayes:
- Strong feature dependencies exist
- Need to capture complex non-linear relationships
- Presence of many irrelevant features
- Outliers in data can significantly affect results

### Algorithm Workflow

1. **Load and preprocess data** → Handle missing values, encode categorical features
2. **Calculate prior probabilities** → P(class) for each class
3. **Calculate likelihoods** → Mean and standard deviation for each feature per class
4. **Make predictions** → Calculate posterior probability for each class
5. **Evaluate** → Compare predictions to actual labels, compute accuracy

### Evaluation Metrics

**Accuracy** — fraction of correct predictions:
$$ \text{Accuracy} = \frac{TP + TN}{TP + TN + FP + FN} $$

**Precision** (per class, one-vs-rest) — of all samples *predicted* as this class, how many are correct?
$$ \text{Precision} = \frac{TP}{TP + FP} $$

**Recall** (per class, one-vs-rest) — of all samples that *truly belong* to this class, how many did we find?
$$ \text{Recall} = \frac{TP}{TP + FN} $$

The following helper functions are provided — they use the functions you implemented above.

```python
# These utility functions are provided for you.

def predict(X, priors, class_stats):
    """Predict classes for all samples in X."""
    return [predict_single(sample, priors, class_stats) for sample in X]

def accuracy_rate(y_true, y_pred):
    """Calculate accuracy as a percentage."""
    if len(y_true) == 0:
        return 0.0
    correct = sum(1 for yt, yp in zip(y_true, y_pred) if int(yt) == int(yp))
    return (correct / len(y_true)) * 100

def precision_rate(y_true, y_pred, positive_class=1):
    """Precision for a specific class (one-vs-rest)."""
    tp = sum(1 for yt, yp in zip(y_true, y_pred) if int(yp) == positive_class and int(yt) == int(yp))
    fp = sum(1 for yt, yp in zip(y_true, y_pred) if int(yp) == positive_class and int(yt) != int(yp))
    return tp / (tp + fp) if (tp + fp) > 0 else 0.0

def recall_rate(y_true, y_pred, positive_class=1):
    """Recall for a specific class (one-vs-rest)."""
    tp = sum(1 for yt, yp in zip(y_true, y_pred) if int(yt) == positive_class and int(yt) == int(yp))
    fn = sum(1 for yt, yp in zip(y_true, y_pred) if int(yt) == positive_class and int(yt) != int(yp))
    return tp / (tp + fn) if (tp + fn) > 0 else 0.0
```
